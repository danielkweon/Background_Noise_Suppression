```
# Step 1, requirements, import the following: 
import IPython
from scipy.io import wavfile
import noisereduce as nr
import soundfile as sf
from noisereduce.generate_noise import band_limited_noise
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
import io
%matplotlib inline
```

```
def read_file(file_in):
    rate, data = wavfile.read(file_in)
    return rate, data
```

```
def data_to_file(data, rate, filename):
    wavfile.write(filename, rate, data)
```

```
def display_graph(file_in):
    rate, data = read_file(file_in)
    fig, ax = plt.subplots(figsize=(20,3))
    ax.plot(data)
```

```
def get_data(file_in):
    rate, data = read_file(file_in)
    return data
```

```
def get_rate(file_in):
    rate, data = read_file(file_in)
    return rate
```

```
def noise_reduce_applied(file_in, noise_file, file_out):
    rate, data = read_file(file_in)
    rate_noise, noise = read_file(noise_file)
    data = data/1.0 # float
    noise = noise/1.0 # float
    # 'prop_decrease' is what extent to decrease the noise, we choose to set to 1.0, eliminating all detected noise
    # 'verbose' should be set to true to display plots of steps of algorithm, default is false
    noise_reduced = nr.reduce_noise(audio_clip=data, noise_clip=noise, prop_decrease=1.0, verbose=False)
    data_to_file(noise_reduced, rate, file_out)
    return str(file_out)
```
